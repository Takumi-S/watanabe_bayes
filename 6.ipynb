```python
import numpy as np
import matplotlib.pyplot as plt

def plot_curve(a, b, x_min, x_max):

  def f(x):
      return np.sqrt(np.maximum(x**3 + a*x + b, 0))

  x = np.arange(x_min, x_max, 0.001)
  y = [f(c) for c in x]

  y_max = max(y)

  plt.xlabel('x')
  plt.ylabel('y')
  plt.xlim(x_min, x_max)
  plt.ylim(-y_max, y_max)
  plt.title(f'a={a}, b={b}')

  plt.plot(x, y)
  plt.plot(x, [-c for c in y])
  plt.axhline(0, color='black', linewidth=0.5)
  plt.axvline(0, color='black', linewidth=0.5)
  plt.legend()
  plt.show()
```

```python
plot_curve(0, 0, -1, 5)
```

```python
plot_curve(-3, 2, -3, 3)
```

```python
model17 = """
data {
  int <lower=1> K;
  int <lower=1> N;
  array[N] vector[2] y;
  real beta;
}

parameters {
  simplex[K] theta;
  array[K] vector[2] mu;
  real <lower=0> sigma;
}

transformed parameters {
  vector[K] log_theta = log(theta);
}

model {
  mu ~ multi_normal(rep_vector(0.0, 2), diag_matrix(rep_vector(1.0, 2)));
  for (n in 1:N) {
    vector[K] lps = log_theta;
    for (k in 1:K)
      lps[k] += multi_normal_lpdf(y[n] | mu[k], diag_matrix(rep_vector(sigma, 2)));
    target += beta * log_sum_exp(lps);
  }
}

generated quantities {
  vector[N] log_lik;
  for (n in 1:N) {
    vector[K] lps = log_theta;
    for (k in 1:K)
      lps[k] += multi_normal_lpdf(y[n] | mu[k], diag_matrix(rep_vector(sigma, 2)));
    log_lik[n] = log_sum_exp(lps);
  }
}
"""
```

```python
def wbic(log_likelihood):
  return -np.mean(np.sum(log_likelihood, axis=1))

b_list = [1, 10, 100, 250]
K = 2

x = []
n = 100
for i in range(n):
  x.append([np.random.normal(-1, 1), np.random.normal(-1, 1)])
for i in range(n, 2 * n):
  x.append([np.random.normal(1, 1), np.random.normal(1, 1)])

WBIC = []

for b in b_list:
  data_list = {'K': K, 'N': len(x), 'y': x, 'beta': b / np.log(n)}

  stan_model = stan.build(model17, data=data_list)
  fit = stan_model.sample(num_samples=100, num_chains=1)
  log_lik = fit['log_lik']

  res_wbic = wbic(log_lik)
  WBIC.append(res_wbic)
```

## 問題48

```python
import math
import numpy as np
import statistics

def g(y: list):
  n = len(y)
  return 0.5 * math.log(2 * math.pi * (n + 2) / (n + 1)) + (n + 1) / (n + 2) / 2 * (1 + (-1 * sum(y) / (n + 1)) ** 2)

def t(y: list):
  n = len(y)
  return 0.5 * math.log(2 * math.pi * (n + 2) / (n + 1)) + (n + 1) / (n + 2) / 2 * statistics.mean((y - sum(y) / (n + 1)) ** 2)
```

```python
y = np.random.normal(loc=0, scale=1, size=100)

# 汎化損失
print(g(y))

# 経験損失
print(t(y))
```
